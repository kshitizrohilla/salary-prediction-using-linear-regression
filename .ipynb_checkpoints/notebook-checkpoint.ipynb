```python
import numpy as np
import pandas as pd
from math import sqrt
import matplotlib.pyplot as plt
%matplotlib inline
%config InlineBackend.figure_format = 'retina'
```

```python
df = pd.read_csv("data/salary-data.csv")
display(df.head())
display(df.dtypes)
```

```python
X = df["years_experience"]
y = df["salary"]

print(X.head())
print(y.head())
```

```python
plt.scatter(X, y, color="blue")

plt.title("Scatter plot between salary and years_experience")
plt.xlabel("Experience (in years)")
plt.ylabel("Salary")

plt.show()
```

The slope of the linear regression line $m$ is given by the formula:

<div style="text-align: center; font-size: 25px;">
$m = \frac{\sum_{i=1}^{n}(X_i - \bar{X})(y_i - \bar{y})}{\sum_{i=1}^{n}(X_i - \bar{X})^2}$
</div>

The intercept of the linear regression line $c$ is given by the formula:

<div style="text-align: center; font-size: 25px;">
$c = \bar{y} - m \cdot \bar{X}$
</div>

```python
y_mean = y.mean()
X_mean = X.mean()

numer, denom = 0, 0

for i in range(len(X)):
    numer += (X[i] - X_mean) * (y[i] - y_mean)
    
for i in range(len(X)):
    denom += ((X[i] - X_mean) ** 2)
    
m = numer / denom
print("Slope =", m)

c = y_mean - m * X_mean
print("Intercept =", c)
```

```python
y_pred = m * X + c

plt.plot(X, y, color="blue")
plt.plot(X, y_pred, color="red")
plt.scatter(X, y, color="blue")

plt.show()
```

The Pearson's correlation coefficient $r$ is given by the formula:

<div style="text-align: center; font-size: 25px;">
$r = \frac{{n \sum{X \cdot y} - \sum{X} \cdot \sum{y}}}{{\sqrt{{[n \sum{X^2} - (\sum{X})^2] \cdot [n \sum{y^2} - (\sum{y})^2]}}}}$
</div>

```python
df["Xy"] = X * y
df["X^2"] = X ** 2
df["y^2"] = y ** 2

sum_Xy = df["Xy"].sum()

sum_X = X.sum()
sum_y = y.sum()

sum_X2 = df["X^2"].sum()
sum_y2 = df["y^2"].sum()

n = len(X)

pd.options.display.float_format = '{:.0f}'.format
display(df.head())

r = (n * sum_Xy - sum_X * sum_y) / sqrt((n * sum_X2 - sum_X ** 2) * (n * sum_y2 - sum_y ** 2))
print("Correlation Coefficient (value of r) =", r)
print("Coefficient of Determination (value of r^2 or the Goodness of Fit) =", r ** 2)
```
